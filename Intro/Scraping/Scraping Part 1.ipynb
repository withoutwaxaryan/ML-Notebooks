```python
url_quotes.format("love")
```

```python
res = requests.get(url_quotes)
res.text
```

```python
# not human readable content, therefore we need to parse/ process it.
BeautifulSoup(res.text)
```

```python
soup = BeautifulSoup(res.text)
```

```python
soup.title
```

```python
soup.p
```

```python
soup.find_all("a")
```

```python
quote_divs = soup.find_all("div", attrs={"class" : "quote"})
```

```python
quote_divs
```

```python
quote_div = quote_divs[0]
quote_div
```

```python
quoteText_div = quote_div.find_next("div", attrs={"class" : "quoteText"})
quoteText_div
```

```python
quoteText_div.text
```

```python
quoteText_div.text.strip()
```

```python
striped = quoteText_div.text.strip()
striped_li = striped.split("\n")
```

```python
quote = striped_li[0]
author = striped_li[-1]
```
