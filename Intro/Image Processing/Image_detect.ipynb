```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# In PIL, it was reading image as RGB, but cv2 reads image as BGR i.e. uska opposite
frame = cv2.imread("leo.jpeg")
plt.imshow(frame)
```

```python
rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

```python
# using Haar cascade algorithm and model
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faces = classifier.detectMultiScale(frame) # will give a list of faces detected
```

```python
face = faces[0]
x, y, w, h = face # get the coordinates of the face area
face = rgb[y: y + h, x: x + w] # part of the image containing the face
plt.imshow(face)
```

```python
# Drawing rectangle of face on original image
# cv2.rectangle?
out = cv2.rectangle(rgb, (x,y), (x + w, y + h), (0,0,255), 4) # blue colour rectangle
plt.imshow(out)
```
