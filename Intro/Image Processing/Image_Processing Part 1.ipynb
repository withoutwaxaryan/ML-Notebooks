```python
# (0,0,0) --> Black , (255,255,255) --> White
# Lets make the layer only red colour
red = arr.copy()
red[:, :, [1,2]] = 0 # cut green and blue parts # when we make it 255, ulta hoga
plt.imshow(red)
```

```python
# converting image to blaack and white
gray = arr.mean(axis=2) # axis 0 is width, axis 1 is height, axis 2 is colour
plt.imshow(gray, cmap = "gray")
```
